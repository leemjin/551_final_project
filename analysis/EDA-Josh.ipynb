```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

plt.rcParams["figure.figsize"] = (12, 9) 
```

```python
# Read in data
spy_data = pd.read_csv("../data/spy_processed.csv")
spy_data = spy_data.iloc[6000:]

# log transform some columns
spy_data["volume_dev"] = np.log(spy_data["volume_dev"])
spy_data["range"] = np.log(spy_data["range"])
spy_data["var"] = np.log(spy_data["var"])
```

## Daily Return of S&P 500 from 2018-current

```python
# Convert date column to datetime 
spy_data["Date"] = pd.to_datetime(spy_data["Date"])

# Plot daily open prices
plt.plot(spy_data.Date, spy_data.ret)
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
```

## Log Volume Deviation

```python
plt.plot(spy_data.Date, spy_data.volume_dev)
plt.xlabel("Date")
plt.ylabel("Log Volume Deviation")
plt.show()
```

## Log Range

```python
plt.plot(spy_data.Date, spy_data.range)
plt.xlabel("Date")
plt.ylabel("Log Range")
plt.show()
```

## Log Variance

```python
plt.plot(spy_data.Date, spy_data["var"])
plt.xlabel("Date")
plt.ylabel("Log Variance")
plt.show()
```

The same trends are apparent with daily low prices as the previous stock prices.

```python
fig = plt.figure(figsize=(18, 8))

sub1 = fig.add_subplot(221)
sub1.plot(spy_data.Date, spy_data["ret"])
sub1.set_title('Returns')

sub2 = fig.add_subplot(222)
sub2.plot(spy_data.Date, spy_data["volume_dev"])
sub2.set_title('Log Volume Deviation')

sub3 = fig.add_subplot(223)
sub3.plot(spy_data.Date, spy_data["range"])
sub3.set_title('Log Range')

sub4 = fig.add_subplot(224)
sub4.plot(spy_data.Date, spy_data["var"])
sub4.set_title("Log Variance")

plt.show()
```
